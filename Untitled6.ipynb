```python
import pandas as pd
import numpy as np
import random
from faker import Faker
from datetime import datetime, timedelta
import time

fake = Faker()

# -------------------------------------------------------------
# Configuration
# -------------------------------------------------------------
N_ROWS = 70000
PROFILE_DISTRIBUTION = {"low": 0.30, "medium": 0.50, "high": 0.20}

profiles = {
    "low": {
        "volume_range": (10000, 100000),
        "volume_change_range": (-30, 10),
        "success_rate_range": (85.0, 95.0),
        "refund_rate_range": (0.01, 0.05),
        "chargeback_rate_mean_std": (1.0, 0.2),
        "users_range": (1, 5),
        "login_freq_range": (0.2, 2.0),
        "feature_adoption_range": (10.0, 40.0),
        "last_txn_days_range": (10, 30),
        "support_tickets_range": (5, 30),
        "resolution_time_hours_range": (24, 72),
        "nps_range": (1, 6),
        "contract_types": ["Monthly", "Monthly", "Annual"],
        "base_churn_prob": 0.30
    },
    "medium": {
        "volume_range": (100000, 750000),
        "volume_change_range": (-10, 15),
        "success_rate_range": (95.0, 99.0),
        "refund_rate_range": (0.005, 0.02),
        "chargeback_rate_mean_std": (0.3, 0.1),
        "users_range": (5, 20),
        "login_freq_range": (2.0, 5.0),
        "feature_adoption_range": (40.0, 75.0),
        "last_txn_days_range": (3, 10),
        "support_tickets_range": (2, 10),
        "resolution_time_hours_range": (6, 24),
        "nps_range": (6, 9),
        "contract_types": ["Annual", "Annual", "Monthly"],
        "base_churn_prob": 0.10
    },
    "high": {
        "volume_range": (750000, 5000000),
        "volume_change_range": (5, 30),
        "success_rate_range": (99.0, 99.9),
        "refund_rate_range": (0.001, 0.005),
        "chargeback_rate_mean_std": (0.05, 0.02),
        "users_range": (20, 200),
        "login_freq_range": (5.0, 10.0),
        "feature_adoption_range": (75.0, 100.0),
        "last_txn_days_range": (1, 2),
        "support_tickets_range": (1, 5),
        "resolution_time_hours_range": (1, 8),
        "nps_range": (8, 10),
        "contract_types": ["Annual"],
        "base_churn_prob": 0.02
    }
}

# -------------------------------------------------------------
# Data Generation
# -------------------------------------------------------------
def generate_profiled_data(n_rows, profile_config):
    data = []
    for _ in range(n_rows):
        total_transaction_volume = round(np.random.uniform(*profile_config["volume_range"]), 2)
        quarterly_volume_change = round(np.random.uniform(*profile_config["volume_change_range"]), 2)
        success_rate_30d = round(np.random.uniform(*profile_config["success_rate_range"]), 2)
        api_error_rate_30d = round(max(0, 100 - success_rate_30d - np.random.uniform(0, 2)), 2)
        total_refund_amount_90d = round(total_transaction_volume * np.random.uniform(*profile_config["refund_rate_range"]), 2)
        chargeback_rate_90d = round(max(0, np.random.normal(*profile_config["chargeback_rate_mean_std"])), 4)
        number_of_users = np.random.randint(*profile_config["users_range"])
        avg_login_freq = round(np.random.uniform(*profile_config["login_freq_range"]), 1)
        feature_adoption_rate = round(np.random.uniform(*profile_config["feature_adoption_range"]), 2)
        time_since_last_transaction_days = np.random.randint(*profile_config["last_txn_days_range"])
        support_ticket_count_90d = np.random.randint(*profile_config["support_tickets_range"])
        avg_resolution_hours = round(np.random.uniform(*profile_config["resolution_time_hours_range"]), 1)
        nps_score = np.random.randint(*profile_config["nps_range"])
        account_age_days = np.random.randint(90, 2000)
        contract_type = random.choice(profile_config["contract_types"])
        renewal_date = (datetime.now() + timedelta(days=np.random.randint(30, 730))).strftime('%Y-%m-%d')
        payment_method_type = random.choice(["Credit Card", "Bank Transfer"])

        churn_prob = profile_config["base_churn_prob"]
        if nps_score <= 4: churn_prob += 0.2
        if api_error_rate_30d > 5: churn_prob += 0.1
        if support_ticket_count_90d > (profile_config["support_tickets_range"][1] * 0.7): churn_prob += 0.1
        churned = 1 if np.random.rand() < churn_prob else 0

        data.append({
            "total_transaction_volume_90d": f"${total_transaction_volume:,.2f}",
            "quarterly_volume_change": f"{quarterly_volume_change}%",
            "success_rate_30d": f"{success_rate_30d}%",
            "api_error_rate_30d": f"{api_error_rate_30d}%",
            "total_refund_amount_90d": f"${total_refund_amount_90d:,.2f}",
            "chargeback_rate_90d": f"{chargeback_rate_90d}%",
            "number_of_users": number_of_users,
            "average_user_login_frequency": f"{avg_login_freq} times/week",
            "feature_adoption_rate": f"{feature_adoption_rate}%",
            "time_since_last_transaction": f"{time_since_last_transaction_days} days",
            "support_ticket_count_90d": support_ticket_count_90d,
            "average_ticket_resolution_time": f"{avg_resolution_hours} hours",
            "nps_score": nps_score,
            "account_age_days": account_age_days,
            "contract_type": contract_type,
            "contract_renewal_date": renewal_date,
            "payment_method_type": payment_method_type,
            "churned": churned
        })
    return data


# -------------------------------------------------------------
# Generate Dataset
# -------------------------------------------------------------
all_data = []
current_id = 1000

for profile_name, pct in PROFILE_DISTRIBUTION.items():
    rows = int(N_ROWS * pct)
    profile_data = generate_profiled_data(rows, profiles[profile_name])
    for row in profile_data:
        row["customer_id"] = f"C{current_id}"
        current_id += 1
    all_data.extend(profile_data)

df = pd.DataFrame(all_data)
df.to_csv("synthetic_customer_dataset_70k.csv", index=False)

print("\n‚úÖ Synthetic dataset created: synthetic_customer_dataset_70k.csv")

# -------------------------------------------------------------
# Preprocessing in Chunks (Cleaning + Memory Optimization)
# -------------------------------------------------------------
def preprocess_chunk(df):
    df = df.copy()
    df['total_transaction_volume_90d'] = df['total_transaction_volume_90d'].replace(r'[$,]', '', regex=True)
    df['total_refund_amount_90d'] = df['total_refund_amount_90d'].replace(r'[$,]', '', regex=True)
    percent_cols = ['quarterly_volume_change','success_rate_30d','api_error_rate_30d','chargeback_rate_90d','feature_adoption_rate']
    for col in percent_cols:
        df[col] = df[col].replace(r'[%]', '', regex=True)
    df['average_user_login_frequency'] = df['average_user_login_frequency'].str.split(' ').str[0]
    df['time_since_last_transaction'] = df['time_since_last_transaction'].str.split(' ').str[0]
    df['average_ticket_resolution_time'] = df['average_ticket_resolution_time'].str.split(' ').str[0]
    df['contract_renewal_date'] = pd.to_datetime(df['contract_renewal_date'])
    return df

optimized_dtypes = {
    'customer_id': 'category', 'contract_type': 'category', 'payment_method_type': 'category',
    'number_of_users': 'int16', 'time_since_last_transaction': 'int16',
    'support_ticket_count_90d': 'int16', 'nps_score': 'int8',
    'account_age_days': 'int16', 'churned': 'int8',
    'total_transaction_volume_90d': 'float32', 'total_refund_amount_90d': 'float32',
    'quarterly_volume_change': 'float32', 'success_rate_30d': 'float32',
    'api_error_rate_30d': 'float32', 'chargeback_rate_90d': 'float32',
    'feature_adoption_rate': 'float32', 'average_user_login_frequency': 'float32',
    'average_ticket_resolution_time': 'float32'
}

processed_chunks = []
for chunk in pd.read_csv("synthetic_customer_dataset_70k.csv", chunksize=10000):
    chunk = preprocess_chunk(chunk).astype(optimized_dtypes)
    processed_chunks.append(chunk)

final_df = pd.concat(processed_chunks)

print("\n‚úÖ Data cleaned and optimized")

# -------------------------------------------------------------
# One-Hot Encoding (Correct Final Step)
# -------------------------------------------------------------
final_df_encoded = pd.get_dummies(final_df, columns=['contract_type', 'payment_method_type'], drop_first=True, dtype=int)

print("\n‚úÖ One-hot encoding complete")
print("\nFinal Dataset Shape:", final_df_encoded.shape)

# Save final ML-ready file
final_df_encoded.to_csv("synthetic_customer_dataset_70k_ML_ready.csv", index=False)

print("\nüéØ FINAL OUTPUT: synthetic_customer_dataset_70k_ML_ready.csv")
```

```python
print("TRAINING AND EVALUATING MODELS...")
print("="*80)

def evaluate_model(model, X_train, X_test, y_train, y_test, model_name):
    """Helper function to train and evaluate a model"""

    # Check for NaN values before training
    if X_train.isnull().sum().sum() > 0:
        print(f"‚ö†Ô∏è WARNING: {model_name} - X_train contains {X_train.isnull().sum().sum()} NaN values")
        # Fill NaN with median for this model
        X_train_filled = X_train.fillna(X_train.median())
        X_test_filled = X_test.fillna(X_test.median())
    else:
        X_train_filled = X_train
        X_test_filled = X_test

    try:
        # Train model
        model.fit(X_train_filled, y_train)

        # Predictions
        y_pred = model.predict(X_test_filled)

        # For probability-based metrics
        if hasattr(model, 'predict_proba'):
            y_pred_proba = model.predict_proba(X_test_filled)[:, 1]
        else:
            y_pred_proba = None

        # Calculate metrics
        accuracy = accuracy_score(y_test, y_pred)
        precision = precision_score(y_test, y_pred, zero_division=0)
        recall = recall_score(y_test, y_pred, zero_division=0)
        f1 = f1_score(y_test, y_pred, zero_division=0)

        if y_pred_proba is not None:
            roc_auc = roc_auc_score(y_test, y_pred_proba)
        else:
            roc_auc = 0.5  # Default for models without probability

        # Cross-validation scores (with NaN handling)
        try:
            cv_scores = cross_val_score(model, X_train_filled, y_train, cv=5, scoring='roc_auc')
            cv_auc_mean = cv_scores.mean()
            cv_auc_std = cv_scores.std()
        except:
            cv_auc_mean = 0.5
            cv_auc_std = 0.0
            print(f"‚ö†Ô∏è Cross-validation failed for {model_name}")

        return {
            'model': model_name,
            'accuracy': accuracy,
            'precision': precision,
            'recall': recall,
            'f1_score': f1,
            'roc_auc': roc_auc,
            'cv_auc_mean': cv_auc_mean,
            'cv_auc_std': cv_auc_std,
            'status': 'Success'
        }

    except Exception as e:
        print(f"‚ùå {model_name} failed: {str(e)[:100]}...")
        return {
            'model': model_name,
            'accuracy': 0,
            'precision': 0,
            'recall': 0,
            'f1_score': 0,
            'roc_auc': 0.5,
            'cv_auc_mean': 0.5,
            'cv_auc_std': 0.0,
            'status': f'Failed: {str(e)[:50]}'
        }

# Store results
results = []

# First, let's ensure all data is clean
print("Ensuring data is clean before training...")
print(f"NaN in X_train_scaled: {X_train_scaled.isnull().sum().sum()}")
print(f"NaN in X_test_scaled: {X_test_scaled.isnull().sum().sum()}")
print(f"NaN in X_train_balanced: {X_train_balanced.isnull().sum().sum()}")
print(f"NaN in y_train_balanced: {y_train_balanced.isnull().sum().sum()}")

# Fill any remaining NaN values with median
X_train_scaled_clean = X_train_scaled.fillna(X_train_scaled.median())
X_test_scaled_clean = X_test_scaled.fillna(X_test_scaled.median())
X_train_balanced_clean = X_train_balanced.fillna(X_train_balanced.median())

print(f"After cleaning - NaN in X_train_scaled: {X_train_scaled_clean.isnull().sum().sum()}")
print(f"After cleaning - NaN in X_test_scaled: {X_test_scaled_clean.isnull().sum().sum()}")
print(f"After cleaning - NaN in X_train_balanced: {X_train_balanced_clean.isnull().sum().sum()}")

# Train regular models (using class weights)
print("\nTraining regular models with class weights...")
for name, model in models.items():
    print(f"Training {name}...")
    result = evaluate_model(model, X_train_scaled_clean, X_test_scaled_clean, y_train, y_test, name)
    results.append(result)

# Train balanced models (using SMOTE data)
print("\nTraining models with balanced data...")
for name, model in balanced_models.items():
    print(f"Training {name}...")
    result = evaluate_model(model, X_train_balanced_clean, X_test_scaled_clean, y_train_balanced, y_test, name)
    results.append(result)

# Convert results to DataFrame
results_df = pd.DataFrame(results)

# Filter out failed models
successful_results = results_df[results_df['status'] == 'Success']
failed_results = results_df[results_df['status'] != 'Success']

if len(successful_results) > 0:
    successful_results = successful_results.sort_values('roc_auc', ascending=False)

    print("\n" + "="*80)
    print("SUCCESSFUL MODEL PERFORMANCE COMPARISON")
    print("="*80)
    print(successful_results.round(4))

    # Display best model
    best_model_name = successful_results.iloc[0]['model']
    best_auc = successful_results.iloc[0]['roc_auc']
    print(f"\nüèÜ BEST MODEL: {best_model_name} with ROC-AUC: {best_auc:.4f}")
else:
    print("‚ùå No models trained successfully!")

if len(failed_results) > 0:
    print("\n" + "="*80)
    print("FAILED MODELS")
    print("="*80)
    print(failed_results[['model', 'status']])

print(f"\nüìä Summary: {len(successful_results)} models successful, {len(failed_results)} models failed")
```

```python
print("DETAILED ANALYSIS OF BEST PERFORMING MODEL...")

# Get the best model info
best_model_info = results_df[results_df['status'] == 'Success'].iloc[0]
best_model_name = best_model_info['model']

print(f"Analyzing best model: {best_model_name}")
print(f"Best model ROC-AUC: {best_model_info['roc_auc']:.4f}")

# Use the cleaned data that was used for training
if '(Balanced)' in best_model_name:
    print("Using balanced cleaned data...")
    X_train_clean = X_train_balanced_clean
    y_train_clean = y_train_balanced
else:
    print("Using regular cleaned data...")
    X_train_clean = X_train_scaled_clean
    y_train_clean = y_train

# Re-train the best model with cleaned data
print(f"Re-training {best_model_name} with cleaned data...")

if 'Logistic Regression' in best_model_name:
    if '(Balanced)' in best_model_name:
        best_model = LogisticRegression(random_state=42, max_iter=1000, solver='liblinear')
    else:
        best_model = LogisticRegression(random_state=42, class_weight='balanced', max_iter=1000, solver='liblinear')

elif 'Random Forest' in best_model_name:
    if '(Balanced)' in best_model_name:
        best_model = RandomForestClassifier(random_state=42, n_estimators=100)
    else:
        best_model = RandomForestClassifier(random_state=42, class_weight='balanced', n_estimators=100)

elif 'Gradient Boosting' in best_model_name:
    best_model = GradientBoostingClassifier(random_state=42, n_estimators=100)

elif 'XGBoost' in best_model_name:
    best_model = xgb.XGBClassifier(random_state=42, eval_metric='logloss', n_estimators=100)

elif 'SVM' in best_model_name:
    if '(Balanced)' in best_model_name:
        best_model = SVC(random_state=42, probability=True, kernel='rbf')
    else:
        best_model = SVC(random_state=42, class_weight='balanced', probability=True, kernel='rbf')

# Train the model with cleaned data
best_model.fit(X_train_clean, y_train_clean)

# Predictions using cleaned test data
y_pred = best_model.predict(X_test_scaled_clean)
y_pred_proba = best_model.predict_proba(X_test_scaled_clean)[:, 1]

# Detailed evaluation
print(f"\nDetailed Performance for: {best_model_name}")
print("="*50)

print("\nClassification Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Predicted No Churn', 'Predicted Churn'],
            yticklabels=['Actual No Churn', 'Actual Churn'])
plt.title(f'Confusion Matrix - {best_model_name}')
plt.show()

# Calculate additional metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, zero_division=0)
recall = recall_score(y_test, y_pred, zero_division=0)
f1 = f1_score(y_test, y_pred, zero_division=0)
roc_auc = roc_auc_score(y_test, y_pred_proba)

print(f"\nDetailed Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-Score: {f1:.4f}")
print(f"ROC-AUC: {roc_auc:.4f}")

# Feature Importance (if available)
if hasattr(best_model, 'feature_importances_'):
    print("\nTop 15 Feature Importance:")
    feature_importance = pd.DataFrame({
        'feature': X_train_clean.columns,
        'importance': best_model.feature_importances_
    }).sort_values('importance', ascending=False)

    print(feature_importance.head(15))

    # Plot feature importance
    plt.figure(figsize=(12, 8))
    sns.barplot(data=feature_importance.head(15), x='importance', y='feature')
    plt.title(f'Top 15 Feature Importance - {best_model_name}')
    plt.tight_layout()
    plt.show()
elif hasattr(best_model, 'coef_'):
    print("\nTop 15 Feature Coefficients (Logistic Regression):")
    feature_importance = pd.DataFrame({
        'feature': X_train_clean.columns,
        'coefficient': best_model.coef_[0]
    }).sort_values('coefficient', key=abs, ascending=False)

    print(feature_importance.head(15))

    # Plot feature coefficients
    plt.figure(figsize=(12, 8))
    sns.barplot(data=feature_importance.head(15), x='coefficient', y='feature')
    plt.title(f'Top 15 Feature Coefficients - {best_model_name}')
    plt.tight_layout()
    plt.show()

# ROC Curve
from sklearn.metrics import roc_curve

fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, linewidth=2, label=f'{best_model_name} (AUC = {roc_auc:.4f})')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
plt.show()

# Precision-Recall Curve
from sklearn.metrics import precision_recall_curve, average_precision_score

precision_vals, recall_vals, _ = precision_recall_curve(y_test, y_pred_proba)
avg_precision = average_precision_score(y_test, y_pred_proba)

plt.figure(figsize=(8, 6))
plt.plot(recall_vals, precision_vals, linewidth=2)
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title(f'Precision-Recall Curve (AP = {avg_precision:.4f})')
plt.grid(True)
plt.show()

print(f"\n‚úÖ Detailed analysis completed for {best_model_name}")
```
